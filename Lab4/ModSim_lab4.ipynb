```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

## **Exteral Influence model**

```python
p=np.arange(0.01,0.06,0.01)
dt=0.01
T=100
Na=1000

t=np.arange(0,T+dt,dt)

for val in p:
  x=np.zeros(len(t))
  x[0]=1
  for i in range(1,len(t)):
    x[i]=x[i-1]+dt*val*(Na-x[i-1])
  
  plt.plot(t,x/1000,label=f'p = {val}',lw=2)
  plt.xlabel('Time')
  plt.ylabel('Number of adopters (in thousands)')
  plt.title('External Influence Model')
  plt.legend()
  plt.grid(True)
```

## **Internal Influence Model**

```python
q=np.array([0.1,0.2,0.3,0.4])
dt=0.01
T=100
Na=1000

t=np.arange(0,T+dt,dt)

for val in q:
  x=np.zeros(len(t))
  x[0]=1
  for i in range(1,len(t)):
    x[i]=x[i-1]+dt*val*(Na-x[i-1])*x[i-1]/Na
  
  plt.plot(t,x/1000,label=f'q = {val}',lw=2)
  plt.xlabel('Time')
  plt.ylabel('Number of adopters (in thousands)')
  plt.title('Internal Influence Model')
  plt.legend()
  plt.grid(True)
```

## **Bass Model**

```python
p=np.arange(0.01,0.05,0.01)
q=np.array([0.1,0.2,0.3,0.4])
dt=0.01
T=100
Na=1000

t=np.arange(0,T+dt,dt)

for valp,valq in zip(p,q):
  x=np.zeros(len(t))
  x[0]=1
  for i in range(1,len(t)):
    x[i]=x[i-1]+dt*(valp+valq*x[i-1]/Na)*(Na-x[i-1])
  
  plt.plot(t,x/1000,label=f'p = {valp}, q = {valq}',lw=2)
  plt.xlabel('Time')
  plt.ylabel('Number of adopters (in thousands)')
  plt.title('Bass model')
  plt.legend()
  plt.grid(True)
```

## **Modified Bass Model**

```python
# p=np.arange(0.01,0.05,0.01)
# q=np.array([0.1,0.2,0.3,0.4])
p=0.01
q=0.1
beta=np.array([0,1,2,3])
dt=0.001
T=200
Na=1000

t=np.arange(0,T+dt,dt)

for valb in beta:
  x=np.zeros(len(t))
  x[0]=1
  for i in range(1,len(t)):
    temp=q*pow(x[i-1]/Na,valb)
    x[i]=x[i-1]+dt*(p+temp)*(Na-x[i-1])
  
  plt.plot(t,x/1000,label=f'\u03B2 = {valb}',lw=2)
  plt.xlabel('Time')
  plt.ylabel('Number of adopters (in thousands)')
  plt.title('Modified Bass model')
  plt.legend()
  plt.grid(True)
```
