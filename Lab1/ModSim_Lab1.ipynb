```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def radioactive_model(iter,A0,a,b,h):
  A=np.zeros(iter)
  B=np.zeros(iter)
  C=np.zeros(iter)

  A[0]=A0

  for i in range(1,iter):
    A[i]=A[i-1]-h*a*A[i-1]
    B[i]=B[i-1]+h*a*A[i-1]-b*h*B[i-1]
    C[i]=C[i-1]+h*b*B[i-1]
  
  return A,B,C
```

```python
a,b=0.02,0.004    #rate of decay
start,end=0,1000
h=0.001
iter=int((end-start)/h)
A0=500

A,B,C=radioactive_model(iter,A0,a,b,h)
```

```python
plt.figure(1)
plt.plot(np.arange(start,end,h),A,'g')
plt.plot(np.arange(start,end,h),B,'b')
plt.plot(np.arange(start,end,h),C,'r')
plt.legend(['Mass of A','Mass of B','Mass of C'])
plt.xlabel('Time')
plt.ylabel('Mass of the Substance')
plt.grid()
```

```python
total_disintegrations=np.zeros(iter)

for i in range(iter):
  total_disintegrations[i]=a*A[i]+b*B[i]

plt.figure(2)
plt.plot(np.arange(start,end,h),total_disintegrations,'g',label='Total disintegrations')
plt.xlabel('Time')
plt.ylabel('Total Disintegrations')
plt.legend()
plt.grid()
```

```python
b=0.3   #rate of decay
start,end=0,100
h=0.001
iter=int((end-start)/h)
A0=500

max_value=[]

for val in a:
  A,B,C=radioactive_model(iter,A0,val,b,h)

  total=np.zeros(iter)
  for j in range(iter):
    total[j]=val*A[j]+b*B[j]
  
  max_value.append(np.argmax(total)/1000)

plt.plot(a,max_value)
plt.xlabel('a')
plt.ylabel('Time of maximum total radioactivity')
plt.grid()
```

```python
a,b=0.1,0.4    #rate of decay
start,end=0,100
h=0.001
iter=int((end-start)/h)
A0=500

A,B,C=radioactive_model(iter,A0,a,b,h)

plt.figure(3)
plt.plot(np.arange(start,end,h),B/A)
plt.xlabel('Time')
plt.ylabel('Mb/Ma')
plt.title('Transient Equilibrium')
plt.grid()
```
